```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('shirt size.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns=['Coller size'])
```

```python
data.head(2)
```

```python
data.head()
```

```python
data = pd.get_dummies(data)
```

```python
data['Recommend CS'] = data['Recommend CS'].astype(str)
```

```python
data.info()
```

```python
data.shape
```

```python
data.head(2)
```

```python
x = data.drop('Recommend CS',axis=1)
y = data['Recommend CS']
```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
score = knn.score(x_test, y_test)
print(score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
score = knn.score(x_test, y_test)
print(score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
pred[0:5]
```

```python
y_test[0:5]
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,pred)
cm
```

```python
import pandas as pd
```

```python
result = pd.DataFrame(data=[y_test.values,pred], index = ['y_test','pred'])
result.transpose()
```

```python
correct_sum = []
for i in range(1,20):
  model = KNeighborsClassifier(n_neighbors=i)
  model.fit(x_train,y_train)
  pred = model.predict(x_test)
  correct = np.sum(pred == y_test)
  correct_sum.append(correct)
```

```python
correct_sum
```

```python
result = pd.DataFrame(data=correct_sum)
result.index = result.index+1
result.T
```
